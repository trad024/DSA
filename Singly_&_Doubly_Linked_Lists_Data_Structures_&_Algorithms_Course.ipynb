##singly linked lists

```python
class SinglyNode:

  def __init__(self,val,next=None):
    self.val=val
    self.next=next
  def __str__(self):
    return str(self.val)
```

```python
Head=SinglyNode(1)
A=SinglyNode(2)
B=SinglyNode(3)
C=SinglyNode(4)
Head.next =A
A.next =B
B.next =C
```

```python
print(A)
```

```python
# Traverse the list - O(n)
curr=Head
while curr:
  print(curr)
  curr=curr.next
```

```python
# Diplay linked list - O(n)
def display(head):
  curr = head
  elements = []
  while curr:
    elements.append(str(curr.val))
    curr = curr.next
  print(' -> '.join(elements))

display(Head)
```

```python
#search for node value - O(n)
def search(head,val):
  curr = head
  while curr:
    if val ==curr.val:
      return True
    curr=curr.next
  return False
```

```python
search(Head,5)
```

# Doubly Linked Lists

```python
class DoublyNode:
  def __init__(self,val,next=None,prev=None):
    self.val=val
    self.next=next
    self.prev=prev
  def __str__(self):
    return str(self.val)
```

```python
head=tail=DoublyNode(1)
print(tail)
```

```python
# Display - O(n)
def display(head):
  curr = head
  elements = []
  while curr:
    elements.append(str(curr.val))
    curr = curr.next
  print(' <-> '.join(elements))

display(head)
```

```python
# Insert at beginning - O(1)
def  insert_at_beginning(head,tail,val):
  new_node=DoublyNode(val,next=head)
  head.prev=new_node
  return new_node, tail
head, tail =insert_at_beginning(head,tail,3)
display(head)
```

```python
# Insert at end - O(1)
def insert_at_end(head, tail, val):
    new_node = DoublyNode(val, prev=tail)
    tail.next = new_node
    return head, new_node

head, tail = insert_at_end(head, tail, 7)
display(head)
```
